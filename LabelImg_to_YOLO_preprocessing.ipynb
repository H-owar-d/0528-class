```python
import os
import random

trainval_percent = 0.1
train_percent = 0.9
txtfilepath = 'labels'
imgfilepath = './JPEGImages/'
total_txt = os.listdir(txtfilepath)

num = len(total_txt)
list = range(num)
tv = max(1, int(num * trainval_percent))
tr = max(1, int(tv * train_percent))
trainval = random.sample(list, tv)
train = random.sample(trainval, tr)

ftrainval = open('trainval.txt', 'w')
ftest = open('test.txt', 'w')
ftrain = open('train.txt', 'w')
fval = open('val.txt', 'w')

for i in list:
    name = imgfilepath + total_txt[i][:-4] + '.jpg\n'
    if i in trainval:
        ftrainval.write(name)
        if i in train:
            ftest.write(name)
        else:
            fval.write(name)
    else:
        ftrain.write(name)


ftrainval.close()
ftrain.close()
fval.close()
ftest.close()
```
